```python
# Function to clean and calculate daily returns for a given DataFrame
def calculate_daily_returns(df, name):
    # Drop the first row which contains descriptive text instead of data
    df = df.iloc[1:].copy()

    # Convert 'Date' to datetime objects and sort
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date').reset_index(drop=True)

    # Clean 'Adj Close' column: remove commas and convert to numeric
    # Using errors='coerce' will turn any unparseable values into NaN
    df['Adj Close'] = df['Adj Close'].astype(str).str.replace(',', '', regex=False)
    df['Adj Close'] = pd.to_numeric(df['Adj Close'], errors='coerce')

    # Calculate daily returns
    df['Daily Return'] = df['Adj Close'].pct_change() * 100

    print(f"\n{name} Daily Returns (first 5 rows with original columns and new 'Daily Return' column):")
    display(df.head())
    return df

# Process each DataFrame
df_bbca = calculate_daily_returns(df_bbca, 'BBCA')
df_bbni = calculate_daily_returns(df_bbni, 'BBNI')
df_bbri = calculate_daily_returns(df_bbri, 'BBRI')
df_bmri = calculate_daily_returns(df_bmri, 'BMRI')
```

```python
import pandas as pd

def calculate_daily_returns(df, name):
    # Strip whitespace from column names
    df.columns = df.columns.str.strip()

    df = df.iloc[1:].copy()

    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by='Date').reset_index(drop=True)

    df['Adj Close'] = df['Adj Close'].astype(str).str.replace(',', '', regex=False)
    df['Adj Close'] = pd.to_numeric(df['Adj Close'], errors='coerce')

    df['Daily Return'] = df['Adj Close'].pct_change() * 100

    print(f"\n{name} Daily Returns (first 5 rows with original columns and new 'Daily Return' column):")
    display(df.head())
    return df

df_bbca = calculate_daily_returns(df_bbca, 'BBCA')
df_bbni = calculate_daily_returns(df_bbni, 'BBNI')
df_bbri = calculate_daily_returns(df_bbri, 'BBRI')
df_bmri = calculate_daily_returns(df_bmri, 'BMRI')
```

```python
import matplotlib.pyplot as plt

def plot_daily_returns(df, name, color):
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['Daily Return'], label=f'{name} Daily Return', color=color)
    plt.title(f'Daily Returns for {name} (2024)')
    plt.xlabel('Date')
    plt.ylabel('Daily Return (%)')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_daily_returns(df_bbca, 'BBCA', 'blue')
```
